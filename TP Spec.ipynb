```python
#Fit Gaussienne:
from scipy.optimize import curve_fit
def gaus(x,a,x0,sigma,h):
    return a*np.exp((-(x-x0)**2)/(2*sigma**2))+h
popt, pcov = curve_fit(gaus,Xpic1,-Pic1,p0=[1,9*10**9,10**9,-0.085])
plt.plot(Xpic1,-Pic1)
plt.plot(Xpic1,gaus(Xpic1,*popt))
print(np.diag(pcov))
```

```python
def gaus2(x,a1,m1,sigma1,a2,m2,sigma2,h):
    return a1*np.exp((-(x-m1)**2)/(2*sigma1**2))+a2*np.exp((-(x-m2)**2)/(2*sigma2**2))+h
Xd=X3[Ni:Kf]*10**(-9)
Picd=-Y3[Ni:Kf]/10
popt, pcov=curve_fit(gaus2,Xd,Picd,p0=[1,9,0.3,0.5,10.25,0.6,0.0085])
print(np.diag(pcov))
plt.plot(Xd,Picd)
plt.plot(Xd,gaus2(Xd,*popt))
```
